### Web Scrappng: saving data into Csv and Excel

<h4>1.Import Libraries

```python
import bs4
import urllib.request
from urllib.request import urlopen
from bs4 import BeautifulSoup as soup
import csv
```

<h4>2.Load the Html Code from url 
<h4>2.I. write to csv files

```python
filename = 'Music_Distributions_Ratings.csv'
f = open(filename, 'w',newline = '')
music = csv.writer(f) #to write into cvs file

html = urlopen('https://en.wikipedia.org/wiki/List_of_largest_recorded_music_markets')
bsobj = soup(html.read())
tbody = bsobj('table',{'class':'wikitable plainrowheaders sortable'})[3].findAll('tr')
tbody
xl = [] #to write into excel file
for row in tbody:
    cols = row.findChildren(recursive = False)
    cols = [element.text.strip() for element in cols]
    music.writerow(cols)
    xl.append(cols)
    print(cols)
```

```python
import os
os.getcwd()
```

3.To write to excel file

```python
xl
```

```python
import pandas as pd
df = pd.DataFrame(data =xl)
df
```

3. Now, let's run it into Excel format.

```python
df.to_excel('Music_Distributions_Ratings.xlsx', index= False,header =False)
print("Spreadsheet saved.")
```
